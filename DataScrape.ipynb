```python
# %pip install lxml
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
## Establish chrome is the browser we use
driver = webdriver.Chrome()
## Root url of the data
url = r"https://www.nba.com/stats/players/boxscores?SeasonType=Regular+Season&Season=2018-19"
driver.get(url)
```

```python
## Make the table to list all data, the xpath here is universal to all season data page
select = Select(driver.find_element("xpath", r'//*[@id="__next"]/div[2]/div[2]/div[3]/section[2]/div/div[2]/div[2]/div[1]/div[3]/div/label/div/select'))
## All data option is the first one
indx = 1
player_data = pd.DataFrame()
for i in np.arange(1,524):
    select.select_by_index(i)
    src = driver.page_source
    parser = BeautifulSoup(src, "lxml")
    table = parser.find("div", attrs = {"class": "Crom_container__C45Ti"})
    headers = table.findAll('th')
    headerlist = [h.text.strip() for h in headers]
    ## This is all of the player data but in nasty format
    rows = table.findAll('tr')[1:]
    new_player_data = pd.DataFrame([[td.getText().strip() for td in rows[i].findAll('td')] for i in range(len(rows))])
    player_data = pd.concat([player_data, new_player_data])
player_data
```

```python
player_data.to_csv('2018_19.csv', sep='\t', encoding='utf-8')
```
